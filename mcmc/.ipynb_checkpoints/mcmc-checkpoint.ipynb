## La Vraisemblance marginale

La vraisemblance marginale mesure à quel point le modèle complet (paramètres + données) explique bien les données observées, en tenant compte de toutes les valeurs possibles du paramètre, pondérées par leur probabilité a priori.

Autrement dit, c’est une moyenne pondérée de la vraisemblance sur toutes les valeurs possibles de $\theta$, l'idée c'est d'avoir un metrique qui exprime à quel point un modèle décrit bien les données indépendament de la valeur que son paramètre peut prendre. 

Formellement cela se calcule comme une intégrale, donc dans notre cas il se formalise comme il suit:

$$ P(x) = \int d\theta \  P(x \mid \theta) \cdot P(\theta) $$



### Exemple de intégration numérique à la main

Dans le chapitre sur les priors, nous avons calculé la distribution a posteriori en utilisant la loi Bêta et sa constante de normalisation.
Ici, nous allons calculer la même distribution a posteriori en utilisant l'intégration numérique.
Nous allons reprendre exactement le même exemple : dix personnes ont suivi le traitement, et trois d’entre elles ont eu une rechute à six mois.
Nous utilisons un modèle basé sur une loi binomiale, avec un paramètre $\theta$ représentant la probabilité de rechute à estimer, et un prior défini par une loi Bêta avec $\alpha = \beta = 2$

```r
# Données
k = 3
n = 10
# Hyperparamètres du prior
alpha = 2
beta = 2
# Espace des paramètres à explorer (variant de 0 à 1 avec des pas de 0,002)
p <- seq(0,1,.002)
```

À partir de cela, nous allons construire une fonction qui, pour chaque valeur du paramètre, multiplie la vraisemblance (calculée avec un modèle binomial) par le prior (défini à l’aide d’une loi Beta).
Cette opération correspond au numérateur dans la formule de Bayes, c’est-à-dire la partie qui combine ce que nous savons avant (le prior) et ce que nous apprenons avec les données (la vraisemblance) et dans le code elle s'implément comme il suit:

```r
# numérateur dans la forumle de bayse : vraisemblance * prior
numerator <- function(p) dbinom(k,n,p)*dbeta(p,alpha,beta)
```

Nous pouvons créer une representation graphique de cette distribution comme il suit:

```r
# mettre le tout dans une dataframe
df <- data.frame(prob = p, num = numerator(p))

# faire le graphique
ggplot(df, aes(x = prob, y = num)) +
  geom_area(fill = "lightblue", alpha = 0.4) +
  geom_line(color = "darkblue", linewidth = 1.2) +
  annotate("text", x = max(df$prob)*0.8, y = max(df$num) * 0.9, 
           label = paste0("Aire sous la courbe = ", round(denominator,3)),
           size = 5, color = "black") +
  labs(
    x = expression(theta),
    y = "Vraisemblance x Prior"
  ) +
   theme_minimal(base_size = 18) 
```

Vous pouvez voir que cette distribution n'est pas une fonction de densité valide, car son intégrale ne correspond pas à 1. Pour obtenir une distribution a posteriori qui représente une fonction de densité valide, il nous faut cette fameuse constante de normalisation $P(D)$.
Pour l'obtenir, nous devons d'abord calculer l'intégrale de cette distribution. Comme il s'agit d'une intégrale simple, nous pouvons faire appel à la fonction `integrate`, comme suit :


```r
# denominateur (aire sous la curbe)
denominator <- integrate(numerator,0,1)$value

```

Ensuite, il suffit de diviser le numérateur (vraisemblance × prior) par la valeur de l'intégrale, comme suit :

```r
# posterior
posterior <- numerator(p)/denominator
```

Si l'on fait une représentation graphique de cette distribution, on peut voir qu'il s'agit bien d'une fonction de densité valide, et que son intégrale est égale à 1 :

```r
# mettre le tout dans une dataframe
df <- data.frame(prob = p, post = posterior)

# calcule de l'intégrale du posterior
post <-function(p) numerator(p)/denominator

aire_curbe <- integrate(post,0,1)$value


# faire le graphique
ggplot(df, aes(x = prob, y = post)) +
  geom_area(fill = "orange", alpha = 0.4) +
  geom_line(color = "darkred", linewidth = 1.2) +
  annotate("text", x = max(df$prob)*0.8, y = max(df$post) * 0.9, 
           label = paste0("Aire sous la curbe = ", round(aire_curbe,3)),
           size = 5, color = "black") +
  labs(
    x = expression(theta),
    y = "Vraisemblance x Prior / intégrale"
  ) +
   theme_minimal(base_size = 18) 
```

Comme confirmation, vous pouvez observer comment cette distribution correspond exactement à ce que nous avons calculé en utilisant la loi Bêta comme constante de normalisation dans le chapitre sur les priors.

```r
# hyper paramètres de la distribution a priori
alpha_pr = 2
beta_pr  = 2

# data
k = 3
n = 10

# hyper paramètres de la distribution a posteriori
alpha <- alpha_pr + k
beta  <- beta_pr + n - k

print(alpha)
print(beta)

# Create a grid of values between 0 and 1
x <- seq(0.0001, 0.9999, by = 0.0001)
y <- dbeta(x, alpha, beta)

#
samples <- rbeta(10000, alpha, beta)

# Data for plotting
df <- data.frame(x = x, y = y)

# Create the plot
ggplot(df, aes(x = x, y = y)) +
  geom_line(color = "darkred", linewidth = 1.2) +
  labs(title = "",
       x = expression(theta),
       y = "Densité") +
  theme_minimal(base_size = 18)
```

## L'approximation de la distribution à posteriori par MCMC


Dans les MCMC on utilise  marche alétoire (*random walk*) our explorer l'espace des paramètres (par exemple, les valeurs possibles de $\theta$). Une marche aléatoire est un processus stochastique (c’est-à-dire basé sur le hasard) dans lequel on se déplace pas à pas, à chaque étape en choisissant une nouvelle position au hasard, autour de la position précédente. 

Cette marche alétoire est dit Markovienne, parce que chaque pas au moment ($t$) dépend du pas prédedant ($t-1$) et non pas des autres pas fait dans le passé (e.g., $t-2$, $t-3$, $t-4$, ...).

### Définir les fonctions de vraisemblance, du prior et du posterior

Nous allons maintenant définir les fonctions de vraisemblance, du prior et du posterior.  
Avant cela, nous devons faire une petite modification  par rapport à ce que nous avons vu jusqu’à présent.

Nous allons travailler avec les logarithmes de la vraisemblance et du prior. Cette transformation permet de remplacer les opérations multiplicatives (souvent utilisées en probabilités) par des opérations additives, ce qui est plus stable numériquement et plus facile à implémenter en `R`. Concrètement cela signifie que nous allons modifer la formule de Bayes classique qui s’écrit normalement ainsi :

$$
P(\theta \mid D) = \frac{P(D \mid \theta) \cdot P(\theta)}{P(D)}
$$

En prenant le logarithme de chaque terme (comme le logarithme d’un produit est la somme des logarithmes), nous allons donc l'écrire comme il suit  :

$$
\log P(\theta \mid D) = \log P(D \mid \theta) + \log P(\theta) - \log P(D)
$$

Cette forme additive est particulièrement utile en programmation, parce que elle évite les erreurs d’arrondi dues à des très petits nombres multipliés. Dans le cas specifique de l'algorithm de Metropolis, cela simplifie les calculs dans la comparison des ratios (que on verra en suite dans les détails).

Commencons par définir la fonction qui nous donnera le log de la vraisemblance, il s'agit toujour d'une binomiale (`dbinom`), mais cette fois on calcule le logarithme de la vraisemblance, en spécifiant `log = TRUE`


```r
# log-likelihood function

loglikelihood <- function(x, p){
  dbinom(x = x, size = n, prob = p, log = TRUE)
}
```

Nous allons faire de même pour le prior.  
Il s'agit toujours d'une loi Bêta avec $ \alpha = \beta = 2 $, transformé en logarithme :


```r
# prior density
logprior <- function(p){
  dbeta(p,2,2, log = TRUE)
}
```

Finalement  ous allons maintenant définir la fonction pour le posterior. Ici, vous pouvez remarquer que la probabilité des données `log(Pr(D))` est mise en commentaire dans le code.  Cela est possible parce que dans l’algorithme de Metropolis, nous allons plutôt comparer des ratios entre différentes valeurs du postérieur. Or, dans ces ratios la probabilité des données $P(D)$  apparaît à la fois au numérateur et au dénominateur, donc il s’annule automatiquement.  C’est un avantage majeur de cet algorithme ( et des MCMC en général): il nous permet de éviter le calcul de l’intégrale. 
Ainsi, nous pouvons travailler avec le logarithme du poterior non normalisé, sans avoir à calculer l’évidence.


```r
# posterior density function (log scale) 
 posterior <- function(x, p){
  loglikelihood(x, p) + logprior(p) # - log(Pr(data)) 
}
```

### Définir la fonction qui propose la nouvelle valeur



```r
# propose candidate value
move <- function(x, away = .2){ 
  logitx <- log(x / (1 - x))
  logit_candidate <- logitx + rnorm(1, 0, away)
  candidate <- plogis(logit_candidate)
  return(candidate)
}
```

[ADD HERE EXPLANTION ABOUT METROPOLIS]

```r
metropolis <- function(steps = 100, inits = 0.5){
  
  # pre-alloc memory
  theta.post <- rep(NA, steps)
  
  # start
  theta.post[1] <- inits
  
  for (t in 2:steps){
    
    # propose candidate value for prob of success
    theta_star <- move(theta.post[t-1])
    
    # calculate ratio R
    pstar <- posterior(k, p = theta_star)  
    pprev <- posterior(k, p = theta.post[t-1])
    logR <- pstar - pprev
    R    <- exp(logR)
    
    # decide to accept candidate value or to keep current value
    accept <- rbinom(1, 1, prob = min(R, 1))
    theta.post[t] <- ifelse(accept == 1, theta_star, theta.post[t-1])
  }
  theta.post
}
```

```r
steps <- 2000
chain1 <- metropolis(steps = steps, inits = 0.2)
chain2 <- metropolis(steps = steps, inits = 0.5)
chain3 <- metropolis(steps = steps, inits = 0.7)

df <- data.frame(iter = rep(1:steps, 3), 
                 value = c(chain1, chain2, chain3),
                 chain = c(rep("chain1", steps), 
                           rep("chain2", steps), 
                           rep("chain3", steps)))

```
