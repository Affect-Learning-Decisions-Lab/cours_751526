### Exemple à la main

Dans le chapitre sur les priors, nous avons calculé la distribution a posteriori en utilisant la loi Bêta et sa constante de normalisation.
Ici, nous allons calculer la même distribution a posteriori en utilisant l'intégration numérique.
Nous allons reprendre exactement le même exemple : dix personnes ont suivi le traitement, et trois d’entre elles ont eu une rechute à six mois.
Nous utilisons un modèle basé sur une loi binomiale, avec un paramètre θθ représentant la probabilité de rechute à estimer, et un prior défini par une loi Bêta avec $\alpha = \beta = 2$

```r
# Données
k = 3
n = 10
# Hyperparamètres du prior
alpha = 2
beta = 2
# Espace des paramètres à explorer (variant de 0 à 1 avec des pas de 0,002)
p <- seq(0,1,.002)
```

À partir de cela, nous allons construire une fonction qui, pour chaque valeur du paramètre, multiplie la vraisemblance (calculée avec un modèle binomial) par le prior (défini à l’aide d’une loi Beta).
Cette opération correspond au numérateur dans la formule de Bayes, c’est-à-dire la partie qui combine ce que nous savons avant (le prior) et ce que nous apprenons avec les données (la vraisemblance) et dans le code elle s'implément comme il suit:

```r
# numérateur dans la forumle de bayse : vraisemblance * prior
numerator <- function(p) dbinom(k,n,p)*dbeta(p,alpha,beta)
```

Nous pouvons créer une representation graphique de cette distribution comme il suit:

```r
# mettre le tout dans une dataframe
df <- data.frame(prob = p, num = numerator(p))

# faire le graphique
ggplot(df, aes(x = prob, y = num)) +
  geom_area(fill = "lightblue", alpha = 0.4) +
  geom_line(color = "darkblue", linewidth = 1.2) +
  annotate("text", x = max(df$prob)*0.8, y = max(df$num) * 0.9, 
           label = paste0("Aire sous la curbe = ", round(denominator,3)),
           size = 5, color = "black") +
  labs(
    x = expression(theta),
    y = "Vraisemblance x Prior"
  ) +
   theme_minimal(base_size = 18) 
```

Nous pouvons voir que cette distribution n'est pas une PDF valide son intégrale ne correspond pas à 1. Pour arriver à un posteriori qui represented une PDF valide il nous faut cette fameuse constante de normalisation P(D). 
Pour l'obtenir on doit d'abord calculer l'intégrale de cette distribution, comme il s'agit d'une intégrale simple on peut fair appel à la functionn`integrate` comme il suit:


```r
# denominateur (aire sous la curbe)
denominator <- integrate(numerator,0,1)$value

```

Et ensuite on peut simplement diviser le numerateur (vraisamblance x prior) par la valeur de l'intégrale, comme il suit:

```r
# posterior
posterior <- numerator(p)/denominator
```

Si on fait un representation graphique de cette distribution on peut voir que elle consiste dans une PDF tout à fait valide et que son intégrale correspond à 1 :

```r
# mettre le tout dans une dataframe
df <- data.frame(prob = p, post = posterior)

# calcule de l'intégrale du posterior
post <-function(p) numerator(p)/denominator

aire_curbe <- integrate(post,0,1)$value


# faire le graphique
ggplot(df, aes(x = prob, y = post)) +
  geom_area(fill = "orange", alpha = 0.4) +
  geom_line(color = "darkred", linewidth = 1.2) +
  annotate("text", x = max(df$prob)*0.8, y = max(df$post) * 0.9, 
           label = paste0("Aire sous la curbe = ", round(aire_curbe,3)),
           size = 5, color = "black") +
  labs(
    x = expression(theta),
    y = "Vraisemblance x Prior / intégrale"
  ) +
   theme_minimal(base_size = 18) 
```

Comme confirmation, vous pouvez voir comment cette distribution correspond exactement à ce que nous avions calculé en utilisant la loi Beta comme constante de normalisation dans la chapitre sur les priors.
